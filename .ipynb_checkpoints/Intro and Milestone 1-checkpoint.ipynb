### 1.1 Import libraries and load data

```python
import pandas as pd
import lightgbm as lgb
import shap
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statistics
from sklearn import metrics
shap.initjs() #This is required to view SHAP plots within a notebook
```

```python
"""Our goal for this task is to interpret a machine learning model using SHAP. The modelling itself is not the main focus of
the task. In the interests of providing an engaging example, the model we are building is a GBM which predicts the 
number of future covid-19 cases per country
from features available in the present.

Based on the default parameters, the target variable is the new daily COVID-19 cases 14 days from the date of feature collection. 

You could pick a different window to 14 days by changing the day_window parameter.

To determine the model's performance, we use a date-based train/test split. You could pick a different date window by changing the training and testing parameters below.

The dataset is pulled from Our World In Data's github repository, so it should be updated daily.

For visualization purposes, we pull out the final day of the testing set and select only countries with fewer than 1000 cases per day.

"""

training_start_date = pd.to_datetime('1 Apr 2020')
training_end_date = pd.to_datetime('15 May 2020')
testing_start_date = pd.to_datetime('15 May 2020')
testing_end_date = pd.to_datetime('1 Jun 2020')
day_window = 14
target_var = 'new_cases_in_%s_days' % day_window
```

```python
def get_latest_covid_data():
    covid_data = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv') #dataset of COVID occurrence/death by country by date
    covid_data = covid_data[~covid_data['location'].isin(['World','International'])] #filter out these special values
    covid_data['date'] = pd.to_datetime(covid_data['date']) #apply datetime transformation
    return covid_data

def join_features_to_target(covid_data, window=day_window):
    #our aim is to predict new covid cases as of 14 days from the date the features are from.
    
    features = covid_data.copy()
    
    target_data = covid_data.copy().rename({'new_cases':target_var}, axis=1) 
    target_data = target_data[['location','date',target_var]]

    features['anchor_date'] = (features['date']+pd.Timedelta(day_window, unit='d')).astype('int64')
    target_data['anchor_date'] = (target_data['date']).astype('int64')
    dataset = pd.merge(features, target_data, on=('location','anchor_date'),how='inner',suffixes=['','_future'])
    dataset = dataset.sort_values(['location','date'])
    return dataset, target_var
    
def train_test_split_by_time(dataset):
    train = dataset[dataset['date'].between(training_start_date,
                                           training_end_date)].set_index(['location','date'])
    test = dataset[dataset['date'].between(testing_start_date,
                                          testing_end_date)].set_index(['location','date'])
    return train, test

covid_data = get_latest_covid_data()
dataset, target_var = join_features_to_target(covid_data)
train, test = train_test_split_by_time(dataset)
visualization = test.reset_index()[(test.reset_index()['date']==test.reset_index()['date'].max()) &
                               (test.reset_index()[target_var]<1000)].set_index(['location','date'])
```

#### 1.2 Investigate missing data, make an initial decision on most useful columns and approach to missingness

```python
#Investigate which columns have high proportions of missing data, and which columns could be transformed from 
#strings variables to numeric variables. Consider which columns you would include in modelling.

#Some columns have high proportions of missing data (>50%). Should we include these? 
#Similarly, should we remove rows with high proportions of missing information?

#The GBM framework we are using (lightgbm) handles missing data intelligently, so in my view it is worth trying to include highly missing columns in initially.
#If the model performs poorly, removing columns with high proportions of missing data can be an idea worth pursuing.

#There is a clearer answer for removing rows. Within the default train period, if we were to filter out rows with missing values
#for new_tests (70% missingness) there are two issues. Firstly, we would lose huge amounts of data. 
#Secondly, we would filter out over 80% of results for countries in Africa, and only 50% for countries in Europe.
#So we risk adding an extra unwanted bias onto our data.  

percent_missing = train.isnull().sum() * 100 / len(train)
missing_value_df = pd.DataFrame({'percent_missing': percent_missing})
print('all rows in train by continent with new_test nulls filtered\n', train[train['new_tests'].isnull()].reset_index()['continent'].value_counts().to_dict())
print('all rows in train by continent\n', train['continent'].value_counts().to_dict())
display(missing_value_df)
```

```python
#Apply the feature columns to your test set
feature_columns = ['population','population_density','median_age','aged_65_older','aged_70_older','gdp_per_capita',
                  'diabetes_prevalence',
                  'hospital_beds_per_thousand',
                  'new_cases',
                  'total_tests',
                  'new_tests',
                  'female_smokers',
                  'male_smokers']
X_train = train[feature_columns]
y_train = train[target_var]
X_test = test[feature_columns]
y_test = test[target_var]
X_visualization = visualization[feature_columns]
y_visualization = visualization[target_var]
```

#### 1.3 Model training

```python
#Train a lightGBM regression to predict the target variable and validate it on the test set using sklearn.metrics.r2_score.
#My r2 score is 0.88 without any hyperparameters tuning. Yours might differ if you chose different parameters.
reg = lgb.LGBMRegressor()
reg.fit(X_train,y_train)
```

```python
metrics.r2_score(y_test,reg.predict(X_test))
```

```python
#useful here to show that our model outperforms just multiplying the new_cases feature by the average growth
y_test.mean() / X_test['new_cases'].mean()
metrics.r2_score(y_test,X_test['new_cases']*1.21)
```

#### 1.4 Create relevant plots, explain model's prediction on one example and on multiple examples

Before we dive into plotting SHAP values are, here is a brief recap from the reading provided above. 

"The goal of SHAP is to explain the prediction of an instance x by computing the contribution of each feature to the prediction." (https://christophm.github.io/interpretable-ml-book/shap.html). 
 - Below you can see a visualization of SHAP values for a single model prediction. 
 - The starting point is the base (expected) value. This is a constant across all rows.
 - From this starting point, the red forces (positive SHAP values) battle the blue forces (negative SHAP values) to push the final prediction up or down respectively. Where they meet is the model's prediction.
 - A positive SHAP value means that for this prediction, the corresponding feature pushes the prediction up by approximately the SHAP value. A negative SHAP value means that for this prediction, the corresponding feature pushes the prediction down.
 - The output value is higher than the base value, which reflects that the positive SHAP values outweigh the negative SHAP values for this row.
 
All global explanations using SHAP are aggregations of these prediction-level local explanations.

!['test'](https://miro.medium.com/max/2000/1*F40hOH_QlubjQuaiqzOPig.png)

```python
"""
Create a SHAP summary plot to understand what the model's reasoning is globally on the train set
 - use shap.TreeExplainer to generate an explanation object for your GBM
 - use explainer.shap_values to generate SHAP values for the training set
 - use shap.summary_plot to generate a plot explaining the model's predictions
 - understand and reflect on the underlying reasoning used by the model, based on the summary plot
 
"""


explainer = shap.TreeExplainer(reg)
shap_values = explainer.shap_values(X_train) #make SHAP explanation values
shap.summary_plot(shap_values, X_train)
```

```python
"""
 - Create a SHAP dependence plot to show how the features 'new_cases' and 'population' interact
 - Is the relationship intuitive? Why do new_cases and population both correlate with high SHAP values for new_cases? 
"""

shap.dependence_plot('new_cases', shap_values, X_train)
```

```python
"""
 - Create a SHAP dependence plot to show how the features 'new_tests' and 'population' interact
 - Does a relationship exist here? How can the numbers of new tests influence cases in the future
 - Is this an example of unwanted bias, where countries that test more have higher predicted cases in the future?
"""

shap.dependence_plot('new_tests', shap_values, X_train)
```

```python
"""
 - Create a SHAP summary plot to understand what the model's reasoning is on a global level for the visualization set
"""

dates_dataset = X_train.reset_index()

shap_values_day = explainer.shap_values(X_visualization) #make SHAP explanation values
shap.summary_plot(shap_values_day, X_visualization)
```

```python
"""
 - Using shap.force_plot on only one row of data, explain the model's reasoning for prediction on that row.
 - In general, the base value is the expected value for the whole training set. If the model's output is lower, it means that the 
 downward pushing forces (blue forces) are stronger than the upward pushing forces (red forces)"""

row_number = 8
shap.force_plot(explainer.expected_value, shap_values_day[row_number,:], X_visualization.iloc[row_number,:]) #explain one row of data
```

```python
"""
 - Use shap.force_plot on the entire visualization set to show the model's reasoning.
 - Note that this force plot, over a whole dataset, is simply a large number of individual force plots which have been flipped and pushed together 
 - Confirm that the underlying story matches the underlying story and important features from the summary plots.
 - In general, the base value is the expected value for the whole training set. If the model's output is lower, it means that the 
 downward pushing forces (blue forces) are stronger than the upward pushing forces (red forces)
 - Explain what patterns shown in this plot are outliers. What are the strongest forces for these outliers?"""

shap.force_plot(explainer.expected_value, shap_values_day, X_visualization) #explain multiple rows of data
```

```python
"""
 - By clicking on a part of the force plot which represents an outlier, obtain its original index. 
 - Then, inspect its individual force plot.
 - Consider if the GBM's reasoning makes sense for this outlier"""

shap.force_plot(explainer.expected_value, shap_values_day[19,:], X_visualization.iloc[19,:]) #explain one row of data
```

```python
"""
 - Use shap.decision_plot to plot SHAP values for the first 20 examples in the visualization set.
 - Often users will find decision_plots easier to interpret than force plots over a whole dataset. They both aim to visualize the same thing, but
 force plots are interactive while decision plots are static"""

shap.decision_plot(explainer.expected_value, shap_values_day[:20], X_visualization[:20]) #explain multiple rows of data
```
